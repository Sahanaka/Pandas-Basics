```python
# Make changes to data
# Get the total of all the featutes and add to a new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head()
```

```python
# Faster method
df["Total"] = df.iloc[:, 4:10].sum(axis=1) # 1 for horizontal addition
df.head()
```

```python
# Change the order of columns
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
df.head()
```

```python
# Saving a new csv
df.to_csv("./DataSets/modified.csv", index=False)
```

```python
# Filtering Data
df.loc[(df['Type 1'] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
new_df = df.loc[(df['Type 1'] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
new_df
```

```python
# Resetting old indexes
new_df.reset_index(inplace=True)
new_df
```

```python
# Contains all the rows with name mega in it
df.loc[df["Name"].str.contains("Mega")]
```

```python
# Drops all the rows with name mega in it
df.loc[~df["Name"].str.contains("Mega")]
```

```python
import re
df.loc[df["Type 1"].str.contains("Fire | Grass", flags=re.I, regex=True)]
```

```python
# Conditional changes
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flame"
df
```
